```python
# Problem 1: Employee Performance Bonus Eligibility

# Dictionary containing employee names and their performance scores
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}
```

```python
# Step 1: Find the highest score
highest_score = max(employees.values())

```

```python
# Step 2: Identify employees who have the highest score (handling ties)
top_performers = [name for name, score in employees.items() if score == highest_score]
```

```python

# Step 3: Display result
print(f"Top Performers Eligible for Bonus: {', '.join(top_performers)} (Score: {highest_score})")
```
