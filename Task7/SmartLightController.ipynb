```python


class SmartLight:
    def __init__(self, light_name):
        """
        Constructor to initialize light name
        Default status is OFF
        """
        self.light_name = light_name
        self.status = "OFF"
    
    def turn_on(self):
        """Method to turn light ON"""
        self.status = "ON"
        print(f"{self.light_name} is {self.status}")
    
    def turn_off(self):
        """Method to turn light OFF"""
        self.status = "OFF"
        print(f"{self.light_name} is {self.status}")


# Sample Usage
light = SmartLight("Bedroom Light")
light.turn_on()
```
